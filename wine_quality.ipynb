```python
# 필요한 패키지 임포트

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기
df = pd.read_csv("wine-quality.csv")

# 데이터 샘플 확인
df.head()
```

```python
# 산점도 행렬 시각화
sns.set(font_scale=1.1) # 폰트 크기 설정
sns.set_style('ticks')  # 축 눈금 설정
sns.pairplot(df, diag_kind='kde') # 상관계수가 1이면 분포로 표시
plt.show()
```

```python
# 공분산 확인
df.cov()
```

```python
# 피어슨 상관계수 확인
df.corr(method='pearson')
```

```python
# 히트맵 시각화
sns.heatmap(df.corr(), cmap='viridis')
```

```python
# clustermap 히트맵 시각화
sns.clustermap(df.corr(),
               annot = True,
               cmap = 'RdYlBu_r',
               vmin = -1, vmax = 1,
               )
```

```python
# 중복 제거 히트맵 시각화

# 매트릭스의 우측 상단을 모두 True인 1로, 하단을 False인 0으로 변환
np.triu(np.ones_like(df.corr()))

# True/False mask 배열로 변환
mask = np.triu(np.ones_like(df.corr(), dtype=bool))

# 히트맵 그래프 생성
fig, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(df.corr(),
            mask = mask,
            vmin = -1,
            vmax = 1,
            annot = True,
            cmap = "RdYlBu_r",
            cbar = True,
            )
ax.set_title('Wine Quality Correlation', pad = 15)
```
